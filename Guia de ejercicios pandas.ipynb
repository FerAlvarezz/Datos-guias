# Ejercicio N°1
####  Dado el registro de notas de los alumnos de la forma(padrón,materia,nota,fecha).Se pide resolver utilizando Pandas:

**A**.Cuál es el promedio de notas.

**B**.Cuál es la nota más alta y la nota más baja registrada durante el año 2019.

**C**.Cuál es el padrón con mayor cantidad de materias aprobadas durante el último cuatrimestre.

**D**.Cuál es la nota promedio.

**E**.Cuál es la nota promedio por padrón.

```python
notas = pd.DataFrame({'padron': [99373, 99373, 99281, 101923, 102000, 101923, 99373, 92002, 98483],
                      'materia':['AM3', 'F3', 'AM3', 'AM3', 'F3', 'Taller', 'Taller', 'AM2', 'F1'],
                      'nota': [2,9,5,8,4,7,8,9,4],
                      'fecha':['2019-8-2', '2019-2-14', '2018-12-3', '2019-8-2', '2019-7-23','2019-7-27','2018-12-14','2017-7-27','2016-12-10']})
```

```python
notas
```

```python
notas.info()
notas['fecha'] = pd.to_datetime(notas['fecha'])
```

**A**.Cuál es el promedio de notas.

```python
notas['nota'].mean()
```

**B**.Cuál es la nota más alta y la nota más baja registrada durante el año 2019.

```python
notas[notas['fecha'].dt.year == 2019]
```

```python
nota_min = notas[notas['fecha'].dt.year == 2019]['nota'].min()
nota_max = notas[notas['fecha'].dt.year == 2019]['nota'].max()
```

```python
nota_min
```

```python
nota_max
```

**C**.Cuál es el padrón con mayor cantidad de materias aprobadas durante el último cuatrimestre.


```python
notas_fec = notas[notas['fecha'] > '2019-3-1']
max_aprob = notas_fec[notas_fec['nota'] >= 4]['padron'].value_counts().idxmax() #seguir preguntando
max_aprob
```

**D**.Cuál es la nota promedio.

```python
notas[notas['padron']==max_aprob]['nota'].mean()
# o me pide notas['nota'].value_counts() ?
```

**E**.Cuál es la nota promedio por padrón.

```python
notas.groupby('padron').agg({'nota':'mean'}).round(1) #round es para redondear.
```

# Ejercicio 4

4- Un sitio de Ebooks tiene información sobre los reviews que los usuarios hacen de sus libros en un DataFrame con formato (user_id, book_id, rating, timestamp). Por otro lado tenemos información en otro DataFrame que bajamos de GoodReads: (book_id, book_name, avg_rating). Podemos suponer que los Ids de los libros son compatibles. Se pide usar Python Pandas para:

    A. Obtener un DataFrame que indique el TOP5 de Ebooks en el sitio de Ebooks. (Para este punto se puede ignorar el segundo DataFrame).
    B. Obtener un DataFrame que indique qué libros tienen una diferencia de rating promedio mayor al 20% entre el sitio de Ebooks y GoodReads.



```python
df_ebooks = pd.DataFrame(columns = ['user_id', 'book_id', 'rating', 'timestamp'])
df_goodreads = pd.DataFrame(columns = ['book_id', 'book_name', 'avg_rating'])
libros = ['Narnia', 'Mi Queso', 'Jesus', 'Data Mining', 'Colores']
for id_libro in range(len(libros)) :
    for j in range(3):
        df_ebooks = df_ebooks.append({
                        'user_id' : j + 200,
                        'book_id' : id_libro,
                        'rating' : random.uniform(0, 100),
                        'timestamp' : random.choice(['2017-8-1','2018-1-2','2019-9-4','2019-9-9'])
                    }, ignore_index = True)
    df_goodreads = df_goodreads.append({
                            'book_id' : id_libro,
                            'book_name' : libros[id_libro],
                            'avg_rating' : random.uniform(0, 100)
                        }, ignore_index = True)
```

```python
df_ebooks.head(2)
```

```python
df_goodreads.head(2)
```

```python
df_ebooks.nlargest(5, 'rating')
```

```python
df_ebooks = df_ebooks.groupby('book_id').agg({'rating':'mean'})
df_ebooks
```

```python
df_goodreads = df_goodreads.groupby('book_id').agg({'avg_rating':'mean'})
df_goodreads
```

```python
merge = pd.merge(df_goodreads, df_ebooks, on='book_id', how='inner').diff(axis=1).drop(columns=['avg_rating']).applymap(lambda x: np.abs(x))
merge[merge['rating'] > 20]
```

5- La Agencia Nacional de Estadísticas de Buenos Aires recolecta información de nacimientos cuando los padres registran a sus hijos en el registro civil a partir de una encuesta. Esa información se encuentra disponible para su análisis en un csv con el siguiente formato (dia_nacimiento, mes_nacimiento, anio_nacimiento, peso_al_nacer, longitud_al_nacer, id_hospital, tipo_parto), donde el tipo de parto 1 es natural y 2 es cesárea. Por otro lado la agencia cuenta con información histórica de los hospitales en otro csv con siguiente formato (id_hospital, dirección, promedio_nacimientos_mensual). Se pide usar Pandas para:

    A. Calcular la cantidad de nacimientos para cada uno de los hospitales para el mes de Octubre de 2017 e indicar aquellos hospitales que superan el promedio de nacimientos mensuales.
    B. Comparando el mes de Octubre de 2017 indicar programáticamente si se incremento el % de cesáreas con respecto a ese mes del año 2016.



```python
df_4_nacimientos_todos = pd.DataFrame(columns = ['dia_nacimiento', 'mes_nacimiento', 'anio_nacimiento','peso_al_nacer','longitud_al_nacer', 'id_hospital', 'tipo_parto'])
df_4_nacimientos_mensuales = pd.DataFrame(columns = ['id_hospital', 'direccion', 'promedio_nacimientos_mensual'])

for id_hospital in range(5):
    for _ in range(50):
        df_4_nacimientos_todos = df_4_nacimientos_todos.append({
                                    'dia_nacimiento' : random.randint(1,28),
                                    'mes_nacimiento' : random.randint(1,12),
                                    'anio_nacimiento' : random.randint(2016,2017),
                                    'peso_al_nacer' : random.uniform(1,5+1),
                                    'longitud_al_nacer' : random.uniform(0.30, 0.50),
                                    'id_hospital' : id_hospital,
                                    'tipo_parto' : random.randint(1,2)
                                }, ignore_index = True)
    df_4_nacimientos_mensuales = df_4_nacimientos_mensuales.append({
                                        'id_hospital' : id_hospital,
                                        'direccion' : ['direccion 1', 'direccion 2', 'direccion 3', 'direccion 4', 'direccion 5'][id_hospital],
                                        'promedio_nacimientos_mensual' : random.randint(1,5) 
                                    }, ignore_index = True)
df_4_nacimientos_todos = df_4_nacimientos_todos.astype({
        'dia_nacimiento' : int, 
        'anio_nacimiento' : int, 
        'mes_nacimiento' : int, 
        'id_hospital' : int, 
        'tipo_parto' : int
})
df_4_nacimientos_mensuales = df_4_nacimientos_mensuales.astype({
        'id_hospital' : int
})
```

Los posibles estados que puede tener un caso son 1: caso abierto, 2: caso resuelto, 3: cerrado sin resolución. Las fechas se encuentran en el formato YYYY-MM-DD.
